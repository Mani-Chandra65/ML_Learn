```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.seasonal import seasonal_decompose

file_path = '/content/IRIS.csv'
iris_data = pd.read_csv(file_path)
```

```python
iris_data
X1 = iris_data[['sepal_length']]
X2 = iris_data[['petal_length']]
y1 = iris_data[['sepal_width']]
y2 = iris_data[['petal_width']]

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

model1=LinearRegression()
model2=LinearRegression()
model1.fit(X1_train,y1_train)
model2.fit(X2_train,y2_train)
```

```python
sepal_length = 6
petal_length = 1.6
input_data = [[sepal_length]]
predict_val=model1.predict(input_data)
print(f"Predicted sepal width:\n{predict_val}")
predict_val=model2.predict(input_data)
print(f"Predicted petal width:\n{predict_val}")
```

```python
plt.scatter(X1,y1,color="Yellow")
plt.plot(X1,model1.predict(X1),color="red")
plt.scatter(X2,y2,color="Blue")
plt.plot(X2,model2.predict(X2),color="green")
plt.title('Length VS Width')
plt.xlabel('Length')
plt.ylabel('Width')
```
