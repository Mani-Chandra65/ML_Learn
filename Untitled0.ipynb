```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
toss=np.array([10,20,30,40,50,60,70,80]).reshape(-1,1)
heads=np.array([5,10,15,20,25,30,35,40])
```

```python
model=LinearRegression()
model.fit(toss,heads)
```

```python
num=340
predict_heads=model.predict([[num]])
print(predict_heads)
```

```python
plt.scatter(toss,heads,color="Yellow")
plt.plot(toss,model.predict(toss),color="red")
plt.title('TOSS VS HEADS')
plt.xlabel('TOSS')
plt.ylabel('HEADS')
```
